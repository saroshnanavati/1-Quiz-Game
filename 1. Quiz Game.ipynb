```python
#Welcome player and setup variables and functions

print("Welcome to my computer quiz!")
playing = input("Do you want to play (y/n)? ").lower()

if playing != "y":
    print('OK, bye now!')
    quit()
else:
    print("OK, let's play!\n")

    score = 0
    count = 0
    count = count + 1

    def report():
        print("\nThat's the end of the quiz!")
        print("You scored ",score," out of ",count," or ",score/count*100,"%")
        quit()
    
    #First Question    

    answer=(input("What does CPU stand for? "))
    if answer.lower() == "central processing unit":
        print ("Correct!")
        score += 1
        if count<4:
            print("Your score is ",score," / ",count)
    else:
        print("Wrong!")
    print("Your score is ",score," / ",count)
    if count<4:
        print("Next Question ...\n")
    else:
        report()

    #Second Question    
        
    count +=  1
    answer=(input("What does GPU stand for? "))
    if answer.lower() == "graphics processing unit":
        print ("Correct!")
        score += 1
        if count<4:
            print("Your score is ",score," / ",count)
    else:
        print("Wrong!")
    print("Your score is ",score," / ",count)
    if count<4:
        print("Next Question ...\n")
    else:
        report()

    #Third Question
        
    count += 1
    answer=(input("What does RAM stand for? "))
    if answer.lower() == "random access memory":
        print ("Correct!")
        score += 1
        if count<4:
            print("Your score is ",score," / ",count)
    else:
        print("Wrong!")
    print("Your score is ",score," / ",count)
    if count<4:
        print("Next Question ...\n")
    else:
        report()

    #Fourth Question
        
    count += 1
    answer=(input("What is 1 + 1? "))
    if answer.lower() == "2":
        print ("Correct!")
        score += 1
        if count<4:
            print("Your score is ",score," / ",count)
    else:
        print("Wrong!")
    print("Your score is ",score," / ",count)
    if count<4:
        print("Next Question ...\n")
    else:
        report()
```
